```python
import requests

# Set headers
headers = requests.utils.default_headers()
headers.update({ 'User-Agent': 'Mozilla/5.0 (X11; Ubuntu; Linux x86_64; rv:52.0) Gecko/20100101 Firefox/52.0'})
```

```python
from bs4 import BeautifulSoup

url = "https://www.billboard.com/charts/year-end/2019/hot-100-songs"
req = requests.get(url, headers)
soup = BeautifulSoup(req.content, 'html.parser')
print(soup.prettify())
```

```python
soup.find_all("article")
```

```python
#The chart list in in section "article", class_="ye-chart-item"

soup.find_all("article", class_="ye-chart-item")
```

```python
# First item of the list, contains rank, artist, title, image, other text

soup.find_all("article", class_="ye-chart-item")[0]
```

```python
# The info we are looking for is in "ye-chart-item__rank", "ye-chart-item__artist" and "ye-chart-item__title"

first_item = soup.find_all("article", class_="ye-chart-item")[0]
```

```python
first_item
```

```python
# The three main pieces of info on every item

rank = first_item.find(class_="ye-chart-item__rank").get_text()
artist = first_item.find(class_="ye-chart-item__artist").get_text()
title = first_item.find(class_="ye-chart-item__title").get_text()

print(rank, artist, title)
```

```python
# The 100 items of the list, called ""ye-chart-item", are in the container "chart-details__item-list"

list = soup.find("div", class_="chart-details")
list
```

```python
#We want to go through all the items in the list

rank_tags = list.select(".chart-details .ye-chart-item__rank")
ranks = [rt.get_text() for rt in rank_tags]
ranks
```

```python
# Removing the "\n" characters
ranks_clean = [x.replace('\n', '') for x in ranks]
ranks_clean
```

```python
artist_tags = list.select(".chart-details .ye-chart-item__artist")
artists = [at.get_text() for at in artist_tags]
artists
```

```python
# Removing the "\n" characters
artists_clean = [x.replace('\n', '') for x in artists]
artists_clean
```

```python
title_tags = list.select(".chart-details .ye-chart-item__title")
titles = [tt.get_text() for tt in title_tags]
titles
```

```python
# Removing the "\n" characters
titles_clean = [x.replace('\n', '') for x in titles]
titles_clean.transpose()
```

```python
import numpy as np
import pandas as pd

# Creating list of lists
year_2019_list = [ranks_clean, artists_clean, titles_clean]

#Converting list of lists to pandas DataFrame

year_2019 = pd.DataFrame(year_2019_list).transpose()
year_2019.columns = ['rank', 'artist', 'title']
year_2019
```
