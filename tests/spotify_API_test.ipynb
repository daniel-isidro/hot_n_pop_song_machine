```python
# Let's remove ' Featuring ' wording in the artist column

year_2019['artist_clean'] = year_2019['artist'].replace(to_replace = 'Featuring |& |X ', value = '', regex = True)
```

```python
# Let's replace '$' by 's' in the title column

year_2019['title_clean'] = year_2019['title'].str.replace('$', 's')
```

```python
track_id_list = []

for i in range(0,100):
    search = sp.search(q='artist:' + year_2019['artist_clean'][i] + ' track:' + year_2019['title_clean'][i], type='track')
    if (search['tracks']['items'] == []):
         track_id_list.append(np.nan)
    else:
        track_id = search['tracks']['items'][0]['id']
        track_id_list.append(track_id)
```

```python
# Doing audio_features for first track

result = [sp.audio_features(x) for x in year_2019['track_id']]
```

```python
# Let's review the output

result[0][0]
```

```python
# Doing list comprehension for iterating all rows in the pd.json_normalize(result) dataframe

df = pd.DataFrame()

for i in range(0,100):
    df = df.append(pd.json_normalize(result[i][0]))
```

```python
df
```

```python
# Let's reset the index column

df.reset_index(drop=True, inplace=True)
```

```python
df
```

```python
year_2019
```

```python
# Merging both dataframes

year_2019 = year_2019.merge(df, left_on='track_id', right_on='id', how='outer')
```
