```python
import requests
from bs4 import BeautifulSoup
import pandas as pd

# Set headers
headers = requests.utils.default_headers()
headers.update({ 'User-Agent': 'Mozilla/5.0 (X11; Ubuntu; Linux x86_64; rv:52.0) Gecko/20100101 Firefox/52.0'})

url = "https://en.wikipedia.org/wiki/Billboard_year-end_top_singles_of_1946"
req = requests.get(url, headers)
soup = BeautifulSoup(req.content, 'html.parser')
print(soup.prettify())
```

```python
soup.find_all("table")
```

```python
#The chart list in in section "article", class_="ye-chart-item"

soup.find_all("table", class_="wikitable sortable")
```

```python
# First item of the list, contains rank, artist, title, image, other text

my_table = soup.find('table',{'class':'wikitable sortable'})
my_table
```

```python
position = []
song = []
artist = []
for row in my_table.find_all('tr'):
    cells = row.find_all('td')
    if len(cells)==3:
        position.append(cells[0].find(text=True))
        song.append(cells[1].find_all(text=True))
        artist.append(cells[2].find_all(text=True))
```
