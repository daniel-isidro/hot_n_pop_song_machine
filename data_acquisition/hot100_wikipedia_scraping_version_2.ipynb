```python
import requests
from bs4 import BeautifulSoup
import pandas as pd

# Set headers
headers = requests.utils.default_headers()
headers.update({ 'User-Agent': 'Mozilla/5.0 (X11; Ubuntu; Linux x86_64; rv:52.0) Gecko/20100101 Firefox/52.0'})
```

```python
def hot100(year):
    
    # Valid 1959-1981
    url = "https://en.wikipedia.org/wiki/Billboard_Year-End_Hot_100_singles_of_" + str(year)
    
#   url = "https://en.wikipedia.org/wiki/Billboard_Year-End_Hot_100_singles_of_1981"
#   url = "https://en.wikipedia.org/wiki/Billboard_Year-End_Hot_100_singles_of_1982"
    req = requests.get(url, headers)
    soup = BeautifulSoup(req.content, 'html.parser')

    my_table = soup.find('table',{'class':'wikitable sortable'})
    
    print(soup.prettify())
    
    song = []
    artist = []
    for row in my_table.find_all('tr'):
        cells = row.find_all('td')
        if len(cells)==3:
            song.append(cells[1].find_all(text=True))
            artist.append(cells[2].find_all(text=True))
            
    # Clean up song names
    song_clean = []
    for i in range(0,len(song)):
        if song[i][0]=='"':
            song_clean.append(song[i][1])
        else:
        # Removing quotes in song titles without links
            song_clean.append(song[i][0][1:-1])
    
    # Clean up artist names
    artist_clean = []
    for i in range(0,len(artist)):
        if len(artist[i])<3:
            artist_clean.append(artist[i][0])
        else:
            # Fixing names with several artists
            artist_long = ''
            for j in range(0,len(artist[i])-1):
                artist_long = artist_long + artist[i][j]
            artist_clean.append(artist_long)

    # Removing '\n' characters
    artist_clean = [x.replace('\n', '') for x in artist_clean]

    # Adding 'year' column
    year_list = [str(year)] * 100
    
    # Creating list of lists
    hot100_list = [year_list, artist_clean, song_clean]
    
    #Converting list of lists to pandas DataFrame
    hot100_df = pd.DataFrame(hot100_list).transpose()
    hot100_df.columns = ['year', 'artist', 'title']
    
    return(hot100_df)
```

```python
year1981 = hot100(1981)
```
